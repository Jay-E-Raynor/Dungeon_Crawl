```python
def slow_print(str):
    """Prints text slower"""
    for letter in str:
        sys.stdout.write(letter)
        sys.stdout.flush()
        time.sleep(0.001)
```

```python
def slow_print_2(str):
    """Prints text slower"""
    for letter in str:
        sys.stdout.write(letter)
        sys.stdout.flush()
        time.sleep(0.01)
```

```python
def slow_print_2(lst):
    """Prints text row one by one (list of list)"""
    for row in lst:
        for string in row:
            sys.stdout.write(string)
            sys.stdout.flush()
            time.sleep(0.1)
```
